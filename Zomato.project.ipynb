## 1.Load the Dataset


```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
#read data into a Dataframe
data=pd.read_csv(r'C:\Users\shrik\Downloads\zomato.csv',index_col=0,header=0)
data.head()
```

## 2.Get Basic Information About the Dataset


```python
data.info
```

## 3.Check the Shape and DataType of the Data


```python
print(data.dtypes) # This is used to display the data types of each column in a DataFrame.
print(data.shape) # This will return the number of rows and columns.
```

## 4.Describe the Data


```python
print(data.describe()) # Provides summary statistics like mean, median, and standard deviation for numerical columns.
```

## 5.Check for Missing Values

```python
data.isnull().sum() # This will display the number of missing values per column.
```

## 6.Find Unique Values in a Column



```python
data['area'].unique() #This shows all unique area in the dataset.

```

## 7.Filter Data Based on a Condition



```python
data[data['Rating'] > 4.5].head(10) # This will return all restaurants with a rating greater than 4.5.

```

## 8.Group Data and Find Average Rating by area

```python
data.groupby('area')['Rating'].mean() # This will calculate the average restaurant rating for each area.
```

## 9.Rename the column

```python
data=data.rename(columns={'rate (out of 5)': 'Rating'}) # Renaming the column using df.rename() 
data
```

## 10.How to Visualize top 10 number of restaurants in different area.

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='area', data=data, order=data['area'].value_counts().head(10).index,color='red')
plt.xticks(rotation=45)
plt.title('Number of Restaurants in Top 10 Cities')
plt.show()
```

##  11.Plot a pie chart of restaurant types.

```python
plt.figure(figsize=(8,8))
data['restaurant name'].value_counts().head(5).plot(kind='pie', autopct='%1.1f%%')
plt.title('Top 5 Restaurant Types')
plt.show()
```

## 12.Find the top 10 most expensive restaurants.

```python
print(data.nlargest(10, 'avg cost (two people)')[['restaurant name', 'avg cost (two people)']])
```

## 13.A histogram to show how restaurant ratings are distributed

```python
plt.figure(figsize=(8,6))
sns.histplot(data['Rating'], bins=10, color='purple', kde=True)
plt.title('Restaurant Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

## 14.How the pie chart to visualize the most popular cuisine types

```python
plt.figure(figsize=(8,8))
data['cuisines type'].value_counts().head(7).plot(kind='pie', autopct='%1.1f%%', cmap='Set3')
plt.title('Top 7 Cuisine Types')
plt.ylabel('')
plt.show()
```
